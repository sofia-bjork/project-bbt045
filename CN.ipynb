# Removing adapters

```python
%%bash

cd /cephyr/NOBACKUP/groups/bbt045_2025/groups/group_Bjork/subset_rawdata_RNA-seq
mkdir -p trimmed  # Create output directory if it doesn't exist

for file in *_1.fastq.gz.subset.fastq.gz; do
  # Extract the base filename (remove _1.fastq.gz.subset.fastq.gz)
  base="${file%%_1.fastq.gz.subset.fastq.gz}"
  
  # Define corresponding R2 file
  file_R2="${base}_2.fastq.gz.subset.fastq.gz"

  # Run Cutadapt for paired-end reads
  cutadapt \
    -a AGATCGGAAGAG -A AGATCGGAAGAG \
    --minimum-length 25 \
    -o trimmed/"${base}_trimmed_R1.fastq.gz" \
    -p trimmed/"${base}_trimmed_R2.fastq.gz" \
    "$file" "$file_R2"
done
```

```python
%%bash

cd /cephyr/NOBACKUP/groups/bbt045_2025/groups/group_Bjork/subset_rawdata_RNA-seq/trimmed
mkdir -p fastqc
fastqc -o fastqc *.fastq.gz

cd /cephyr/NOBACKUP/groups/bbt045_2025/groups/group_Bjork/subset_rawdata_RNA-seq/trimmed/fastqc
multiqc .
```

# Test to create count per gene table from bowtie2

```python
%%bash

# Set variables
gtf_file="/cephyr/NOBACKUP/groups/bbt045_2025/groups/group_Bjork/human_genome/Homo_sapiens.GRCh38.113.gtf" #!!!!!!
output_dir="/cephyr/NOBACKUP/groups/bbt045_2025/groups/group_Bjork/data/CN_gene_counts_bowtie2_TEST"
bam_file="/cephyr/NOBACKUP/groups/bbt045_2025/groups/group_Bjork/alignment_Bowtie2_subset_old/ERR950158_trimmed.bam" #!!!!!!
output_file="$output_dir/ERR950158_htseq_counts.txt"

# Create output directory if it does not exist
mkdir -p "$output_dir"

# Sort BAM file by name for htseq-count
sorted_bam_file="$output_dir/ERR950158_sorted.bam"
samtools sort -n -o "$sorted_bam_file" "$bam_file"

# Run htseq-count
htseq-count -f bam -r name -s no -t exon -i gene_id "$sorted_bam_file" "$gtf_file" > "$output_file" 2> "$output_dir/htseq-count.log"

# Check if htseq-count ran successfully
if [ $? -eq 0 ]; then
    echo "htseq-count completed successfully. Output saved in $output_file"
else
    echo "htseq-count failed. Check log file at $output_dir/htseq-count.log" >&2
    exit 1
fi
```
